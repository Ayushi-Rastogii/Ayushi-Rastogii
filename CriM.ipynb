Barnett Reduction

```python
def reduction(u,n):
    q=u//n
    return u-q*n
```

```python
def Barnett_reduction(u,R,N):
  q=int(u*R)>>2*N
  u-=q*N
  while u>=N:
    u-=N
  return u
```

```python
def Barnett_actual_Reduction(u,R,N):
  q=(int(u)>>N-1)*(int(R)>>N+1)
  r1=u%2**(N+1)
  r2=(q*N)%2**(N+1)
  r=r1-r2
  if r<0:
    r=r+2**(N+1)
  while r>=N:
    r-=N
  return r

```

```python
a=100
b=112
n=250
u=a*b
print(reduction(u,n))
R=2**6/250
print(Barnett_reduction(u,R,n))
print(Barnett_actual_Reduction(u,R,n))
```
